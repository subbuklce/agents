Working python objects storing as binary format and JSON format\
pickling - stores the data in binary form so very efficient, python specific only\
josn - supports basic data types, writes class instances as json format to files as well

```python
import pickle as pkl
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __repr__(self):
        return f"Person name is {self.name} and age is {self.age}"
    
person1 = Person("John", 30)
person2 = Person("Jane", 25)

with open("person.pkl", "wb") as file:
    pkl.dump(person1, file)
    pkl.dump(person2, file)

with open("person.pkl", "rb") as file:
    p1= pkl.load(file)
    p2 = pkl.load(file)
    print(p1)
    print(p2)

```

```python
class Book:
    def __init__(self,title:str, author:str):
        self.title = title
        self.author = author
    
    def __str__(self):
        return f"The title is {self.title} and author is {self.author}"

    def get_description(self):
        return (f" The title is '{self.title}' by {self.author}")

ins = Book("Deep Work", "cal")
ins.get_description()

class eBook(Book):
    def __init__(self, title, author, format):
        super().__init__(title, author)
        self.format = format

    def get_description(self):
        return super().get_description() + f"format is {self.format}"

ins1 = eBook("Deep Work", "Cal", "epub")
ins1.get_description()
```

```python
#Explaining the classmethod decorator with inheritance example
from typing import List

class libraryData:
    def __init__(self, title:str, year:int):
        self.title = title
        self.year = year
    def get_description(self):
        return f"'{self.title}' written by {self.author}"
        
class Book(libraryData):
    def __init__(self, title: str, year: int, author:str):
        super().__init__(title, year)
        self.author = author
    def get_description(self):
        return f"'{self.title}' written by {self.author} in {self.year}"
    @classmethod
    def create_multiple(cls, title_authors:List[tuple])->List['Book']:
        return [cls(title, year, author) for title, author, year in title_authors]

books = Book.create_multiple([("Deep Work", 2004, "Cal"),("Let's talk Money", 2022, "Monika Halan")])
for book in books:
    print(book.get_description())

```
